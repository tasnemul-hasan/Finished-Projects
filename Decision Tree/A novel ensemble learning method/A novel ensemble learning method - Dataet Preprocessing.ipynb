# Cancer Dataset

```python
raw_df = pd.read_csv(dataset_path + 'brest_cancer_data.csv')
raw_df
```

```python
raw_df = raw_df.dropna()
raw_df.info()

columns = raw_df.columns.tolist()
count = raw_df[columns].nunique()
print('\n\nNumber of Unique values for Each Columns :')
count
```

```python
header = columns
label_names = 'diagnosis'
X = raw_df.drop(label_names, axis=1)
X.info()
Y = raw_df[label_names]
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42, stratify=Y)
X_train, X_validation, Y_train, Y_validation = train_test_split(X_train, Y_train, test_size=0.2, random_state=42, stratify=Y_train)

# Balancing Training Data
oversample = SMOTE()
X_train, Y_train = oversample.fit_resample(X_train, Y_train)

print('Train: \n', Counter(Y_train))
print('Validation: \n', Counter(Y_validation))
print('Test: \n', Counter(Y_test))

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_validation = scaler.transform (X_validation)
X_test = scaler.transform (X_test)

print('\n')
print('Train: ', X_train.shape)
print('Validation: ', X_validation.shape)
print('Test: ', X_test.shape)

X_train = list(X_train)
X_validation = list(X_validation)
X_test = list(X_test)

Y_train = list(Y_train)
Y_validation = list(Y_validation)
Y_test = list(Y_test)

columns.remove(label_names)
columns.append(label_names)
print('\n\n All the Columns : ', columns)
```

```python
data = {
    'X_train': X_train,
    'X_validation': X_validation,
    'X_test' : X_test,
    'Y_train' : Y_train,
    'Y_validation' : Y_validation,
    'Y_test' : Y_test,
    'header' : header,
    'label_names' : label_names
      }

path = '/content/drive/MyDrive/Colab Notebooks/Decision Tree/Datasets/Datasets/Pickle Formats/'
file_name = 'Cancer'


with open(path+file_name + '.pkl', 'wb') as f:
    pickle.dump(data, f)
```

# Synthetic Dataset


```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from collections import Counter
import pandas as pd
import numpy as np

intensity = ['low', 'moderate', 'high']
types = ['Dim', 'Noise', 'Outlier']
create_dataset = [1, 0, 0]               # create dataset for dimensionality/noise/outlier

dimensionality = [3, 5, 7]
noises = [0.01, 0.1, 0.3]
outliers = [0, 5, 10]

rows = 100

dataset_list = []
# Create a DataFrame
for dim in dimensionality:
  for noise in noises:
    for outlier in outliers:
      X, y = make_classification(
          n_samples = rows + 50,          # Slightly more to ensure balance after random sampling
          n_features = dim,
          n_informative = dim,
          n_redundant = 0,
          n_clusters_per_class = 1,
          flip_y = noise,                 # Low noise
          class_sep = 1.5,
          weights = [0.5, 0.5],           # Trying to maintain balance
          random_state = 42
      )

      # Creating DataFrame
      df_synthetic = pd.DataFrame(X, columns=[f'Feature_{i+1}' for i in range(dim)])
      df_synthetic['Target'] = y

      # Manually balancing to ensure 50 samples of each class
      df_class_0 = df_synthetic[df_synthetic['Target'] == 0].sample(int(rows/2), random_state=42)
      df_class_1 = df_synthetic[df_synthetic['Target'] == 1].sample(int(rows/2), random_state=42)

      # Combining and shuffling to create a perfectly balanced dataset
      df_final_balanced = pd.concat([df_class_0, df_class_1]).sample(frac=1, random_state=42).reset_index(drop=True)

      # Checking the class balance
      final_class_counts = df_final_balanced['Target'].value_counts()
      final_class_counts, df_final_balanced.head()

      #=================================================================================================================================
      # Add outliers
      num_outliers = outlier
      # Select random indices to turn into outliers
      outlier_indices = np.random.choice(df_final_balanced.index, size=num_outliers, replace=False)

      # Add extreme values to these selected rows to create outliers
      df_final_balanced.loc[outlier_indices, df_final_balanced.columns[:-1]] += np.random.uniform(low=-10, high=10, size=(num_outliers, df_final_balanced.shape[1] - 1))

      # Step 3: Shuffle the dataset to randomize the rows
      df_final = df_final_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

      #=================================================================================================================================
      df = df_final.copy()
      df.info()
      columns = df.columns.tolist()
      count = df[columns].nunique()
      print('\n\nNumber of Unique values for Each Columns : \n', count)

      #=================================================================================================================================
      header = columns
      label_names = 'Target'
      X = df.drop(label_names, axis=1)
      X.info()
      Y = df[label_names]
      print(Counter(Y))

      #=================================================================================================================================
      X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42, stratify=Y)
      X_train, X_validation, Y_train, Y_validation = train_test_split(X_train, Y_train, test_size=0.2, random_state=42, stratify=Y_train)


      print('Train: \n', Counter(Y_train))
      print('Validation: \n', Counter(Y_validation))
      print('Test: \n', Counter(Y_test))

      scaler = MinMaxScaler()
      X_train = scaler.fit_transform(X_train)
      X_validation = scaler.transform (X_validation)
      X_test = scaler.transform (X_test)

      print('\n')
      print('Train: ', X_train.shape)
      print('Validation: ', X_validation.shape)
      print('Test: ', X_test.shape)

      X_train = list(X_train)
      X_validation = list(X_validation)
      X_test = list(X_test)

      Y_train = list(Y_train)
      Y_validation = list(Y_validation)
      Y_test = list(Y_test)

      #columns.remove(label_names)
      #columns.append(label_names)
      print('\n\n All the Columns : ', columns)

      #=================================================================================================================================
      data = {
          'X_train': X_train,
          'X_validation': X_validation,
          'X_test' : X_test,
          'Y_train' : Y_train,
          'Y_validation' : Y_validation,
          'Y_test' : Y_test,
          'header' : header,
          'label_names' : label_names
            }

      dataset_list.append(data)
      path = '/content/drive/MyDrive/Colab/Decision Tree/Datasets/Datasets/Pickle Formats/'
      type_index = [index for index, value in enumerate(create_dataset) if value == 1][0]
      if type_index == 0:
        get_intensity = intensity[dimensionality.index(dim)]
      elif type_index == 1:
        get_intensity = intensity[noises.index(noise)]
      elif type_index == 2:
        get_intensity = intensity[outliers.index(outlier)]


      file_name = f'{types[type_index]}_{get_intensity}'
      print(file_name)
      print('\n\n============================\n\n\n')
      with open(path + file_name + '.pkl', 'wb') as f:
          pickle.dump(data, f)

      if create_dataset[2] == 0:
        break

    if create_dataset[1] == 0:
        break

  if create_dataset[0] == 0:
    break



```
